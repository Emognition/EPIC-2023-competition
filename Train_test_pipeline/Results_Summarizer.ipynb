# scenario 1

```python
dfs = []
for scen_1_file in glob.glob('scenario_1*arousal*'):
    df = pd.read_csv(scen_1_file)
    df = df.groupby('x_col_type').mean()
    df['model']=scen_1_file.split('1_')[1].split('across')[0]
    df['file_name']=scen_1_file.split('_')[-1]
    dfs.append(df)
pd.concat(dfs).sort_values('rmses')
```
